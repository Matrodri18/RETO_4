#¿Desea determinar un número real es multiplo de otro número real?

```python
multiploa : float
multiploa = float(input("Ingrese el múltiplo a identificar"))
numerob : float
numerob = float(input("Ingrese un número"))
if (multiploa) % (numerob) == 0:
 print("El número "+ str(multiploa) +" es múltiplo de "+ str(numerob))
else:
  print("El número "+ str(multiploa) +" no es múltiplo de "+ str(numerob))
```
