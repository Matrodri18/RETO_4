#Dado el centro y el radio de un círculo, determinar si un punto de R2 pertenece o no al interior del círculo.

```python
x : float
y : float
h : float
k : float
r : float
r = float(input("ingrese un valor para el radio "))
x = float(input("ingrese un valor para x "))
y = float(input("ingrese un valor para y "))
h = float(input("ingrese un valor para el centro en x "))
k = float(input("ingrese un valor para el centro en y "))
if (x-h)**2 + (y-k)**2 <= r*2:
  print("el punto si se encuentra dentro del circulo")
else:
  print("el punto no se encuentra dentro del circulo")
```
