```python
# ingresar una cadena de caracteres de longitud 1
cadena = input("Ingrese una cadena de longitud 1: ")
# Obtener el código ASCII del carácter
ascii = ord(cadena)
# Determinar si el código ASCII es par o impar
if ascii % 2 == 0:
        print( "El código ASCII de "  + str(cadena) + " es "  +  str(ascii) + " y es par.")
else:
        print( "El código ASCII de "  + str(cadena) + " es " + str(ascii) + " y es impar.")
```
