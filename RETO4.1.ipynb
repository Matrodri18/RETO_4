¿Como saber el coddigo ASCII de una letra minúscula?

```python
n : int
n = int(input("ingrese un numero entero: ")) #conversion a entero
#caracteres correspondientes a los ASCII de vocales minusculas
if n == 97:
  print("El número "  + str(n) + " corresponde al codigo ASCII de la vocal minúscula a")
elif n == 101:
  print("El número "  + str(n) + " corresponde al codigo ASCII de la vocal minúscula e")
elif n == 105:
  print("El número "  + str(n) + " corresponde al codigo ASCII de la vocal minúscula i")
elif n == 111:
  print("El número "  + str(n) + " corresponde al codigo ASCII de la vocal minúscula o")
elif n == 117:
  print("El número "  + str(n) + " corresponde al codigo ASCII la una vocal minúscula u")
else:
    print("El número " +  str(n)  + " no corresponde  al coddigo ASSCI de una vocal minúscula")

```
