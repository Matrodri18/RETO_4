```python
ladoA : float
ladoB : float
hipotenusa : float

ladoA = float(input("ingrese la longitud del lado 1: "))
ladoB = float(input("ingrese la longitud del lado 2: "))
hipotenusa = float(input("ingrese la longitud de la hipotenusa: "))

if ladoA + hipotenusa > ladoB and ladoB + hipotenusa > ladoA and ladoA + ladoB > hipotenusa:
  print("si se puede construir un triangulo con las medidas dadas")
else:
  print("no se puede construir un triangulo, porfavor cambie las medidas4")
```
